```python
import random
developers=list(range(1,51))
testers=list(range(51,81))
managers=list(range(81,100))
sample_dev=random.sample(developers,5)
sample_test=random.sample(testers,3)
sample_mrg=random.sample(managers,2)
final_sample=sample_dev+sample_test+sample_mrg
print(final_sample)

              


```

```python
clusters={
    "Hyderabad":[1,2,3,4,5],
    "Pune":[6,7,8,9],
    "Chennai":[11,12,13,14,15]
}
select_clusters=["Hyderabad","Chennai"]
sample=[]
for cluster in select_clusters:
    sample.extend(clusters[cluster])
    print("clusters sample:",sample)
    
```
