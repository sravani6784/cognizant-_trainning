```python
import plotly.express as px
import pandas as pd
df=pd.DataFrame({
    'months':['jan','feb','mar','apr','may'],
    'sales':[100,200,300,150,300]})
fig=px.line(df,x='months',y='sales',markers=True)
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
months=['jan','feb','mar','may']
sales=[100,200,150,400]
sns.lineplot(x=months,y=sales)
plt.xlabel("months")
plt.ylabel("sales")
plt.title("monthly sales Trends")
plt.show()
```

```python
import matplotlib.pyplot as plt
hours=[1,2,3,4,5]
scores=[50,55,65,67,89]
plt.scatter(hours,scores)
plt.xlabel("hours studies")
plt.ylabel("exam score")
plt.title("score plot exam")
plt.show()

```
