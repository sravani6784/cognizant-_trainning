```python
import matplotlib.pyplot as plt
import numpy as np
products=["mobile","laptop","tv"]
sales_2024=[4000,350,678]
sales_2025=[2000,850,674]
x=np.arange(len(products))
width=0.35
plt.bar(x-width/2,sales_2024,width,label="2024")
plt.bar(x-width/2,sales_2025,width,label="2025")
plt.xlabel("products")
plt.xlabel("sales")
plt.title("sales comparison: 2024 vs 2025")
plt.legend()
plt.show()
        
```

```python
import matplotlib.pyplot as plt
import numpy as np
products=["mobile","laptop","tv"]
sales_2024=[408,874,964]
sales_2025=[584,85,64]
sales_2026=[5400,500,879]
x=np.arange(len(products))
width=0.35
plt.bar(x-width/3,sales_2024,width,label="2024")
plt.bar(x+width/3,sales_2025,width,label="2025")
plt.bar(x+width/2,sales_2026,width,label="2026")
plt.xlabel("products")
plt.xlabel("sales")
plt.title("sales comparison: 2024 vs 2025")
plt.legend()
plt.show()
        
```
