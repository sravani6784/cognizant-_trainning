```python
import numpy as np
salary=([50000,60000,80000,70000,40000])
print(np.median(salary))
print(np.max(salary))
print(np.min(salary))
range=max(salary)-min(salary)
print(range)
print(np.var(salary))
print(np.std(salary))
```

```python
import numpy as np
from scipy import stats
marks=([98,58,97,84,96,74,73])
print(np.mean(marks))
print(np.median(marks))
print(stats.mode(marks))
print(np.max(marks))
print(np.min(marks))
print(np.max(marks)-np.min(marks))
print(np.var(marks))
print(np.std(marks))
```
