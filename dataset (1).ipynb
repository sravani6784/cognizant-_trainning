```python
import pandas as pd
Emp={
          "Name":["Akhilesh","Mahesh","Arun"],
            "salary":[4000,3000,6000]
 } 
result=pd.DataFrame(Emp)
print(result)
print(result[result["salary"]>2000])
print(result.query("salary>2000"))
print(result.query("salary>2000 and salary<3000"))
```
